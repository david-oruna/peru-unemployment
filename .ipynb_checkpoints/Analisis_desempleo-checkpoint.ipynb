```python
# Importamos librerías necesarias
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Leemos y observamos el primer dataset
desempleo_anual = pd.read_csv("Desempleo Anual.csv")
desempleo_anual
```

```python
# Como observamos, la primera fila debería ser las etiquetas de las columnas 
desempleo_anual.loc[0:1]
```

```python
# De todas maneras, la mejor opción es borrar la primera fila y cambiar los nombres de las columnas
desempleo_anual.columns = ['Año','Tasa desempleo']
desempleo_anual.drop(0, axis=0,inplace=True)
```

```python
desempleo_anual
```

```python
# Verificamos nulos
desempleo_anual.isna().sum()
```

```python
# Transformamos la columna años a integers
desempleo_anual['Año']= desempleo_anual['Año'].astype(int)
desempleo_anual['Tasa desempleo'] = desempleo_anual['Tasa desempleo'].astype(float)
```

```python
desempleo_anual
```

```python

# Seleccionamos los años a partir del 2001

desempleo_anual = desempleo_anual[desempleo_anual["Año"]>2000]

# Ahora podemos hacer nuestro primer gráfico

plt.figure(figsize=(10, 6))
plt.bar(desempleo_anual['Año'], desempleo_anual['Tasa desempleo'], color="skyblue")
plt.xlabel('Año')
plt.ylabel('Tasa de desempleo (%)')
plt.title('Tasa de desempleo anual (primeros 10 años)')
plt.tight_layout()
plt.show()
```

Como observamos, ha habido significantes cambios en la tasa de desempleo en los últimos 20 años
Por eso analizaremos ahora la evolución de este tomando en cuenta el género y la edad.


```python
# Ahora veamos los otro datasets
df = pd.read_csv('Desempleo 14_24.csv', encoding='latin1')
df.columns = ['año', 'tasa']
df.drop(0, axis=0, inplace=True)
df.head(10)
```

```python
# Como observamos las fechas están en el formato "mmmYY"
# Por lo que tenemos que convertirlo a fecha
# Para esto usaremos pd.to_datetime; sin embargo, el formato solo acepta meses en inglés

month_translation = {
        'Ene': 'Jan', 'Feb': 'Feb', 'Mar': 'Mar', 'Abr': 'Apr', 'May': 'May', 'Jun': 'Jun', 
        'Jul': 'Jul', 'Ago': 'Aug', 'Sep': 'Sep', 'Oct': 'Oct', 'Nov': 'Nov', 'Dic': 'Dec'
    }
for es, en in month_translation.items():
    df['año']=df['año'].str.replace(es,en)


df['año'] = pd.to_datetime(df['año'], format="%b%y")
df.rename(columns={'año':'fecha'})
```

```python
#Ahora apliquemos lo mismo para los otros dataframes.

def cargar(paths):
    month_translation = {
        'Ene': 'Jan', 'Feb': 'Feb', 'Mar': 'Mar', 'Abr': 'Apr', 'May': 'May', 'Jun': 'Jun', 
        'Jul': 'Jul', 'Ago': 'Aug', 'Sep': 'Sep', 'Oct': 'Oct', 'Nov': 'Nov', 'Dic': 'Dec'
    }
    dfs = []
    for archivo in paths:
        df = pd.read_csv(archivo, encoding='latin1')
        df.columns = ['año', 'tasa']
        df.drop(0, axis=0, inplace=True)
        
        # Traducir nombres de meses
        for es, en in month_translation.items():
            df['año'] = df['año'].str.replace(es, en)
        
        # Convertir a datetime y extraer el año
        df['año'] = pd.to_datetime(df['año'], format='%b%y')
        df.rename(columns={'año':'fecha'})
        
        df.dropna(axis=1)
    
        # Convertir los valores de la columna 'tasa' a numérico, convirtiendo los no numéricos a NaN
        df['tasa']= pd.to_numeric(df['tasa'], errors="coerce").round(2)
        

        
        dfs.append(df)
    
    return dfs

archivos = [
    "Desempleo 14_24.csv", 
    "Desempleo 25_44.csv", 
    "Desempleo 45.csv",
    "Desempleo Trimestral.csv",
    "Desempleo_H.csv",
    "Desempleo_M.csv"
]

dfs = cargar(archivos)


```

```python
desempleo_m = dfs[5]
desempleo_h = dfs[4]
desempleo_m, desempleo_h
```

```python
# Veamos ahora como ha ido variando la tasa de desempleo desde el 2000, según el género.

plt.figure(figsize=(14, 8))

# Plotting men data
plt.plot(desempleo_h['año'], desempleo_h['tasa'], label='Hombres', color='skyblue', marker='o')

# Plotting women data
plt.plot(desempleo_m['año'], desempleo_m['tasa'], label='Mujeres', color='pink', marker='o')

# Encontrar y anotar los puntos más altos
max_h = desempleo_h.loc[desempleo_h['tasa'].idxmax()]
max_m = desempleo_m.loc[desempleo_m['tasa'].idxmax()]

# Anotaciones en los puntos más altos
plt.annotate(f'{max_h["tasa"]}%', (max_h['año'], max_h['tasa']),
             textcoords="offset points", xytext=(0,10), ha='center', color='blue')
plt.annotate(f'{max_m["tasa"]}%', (max_m['año'], max_m['tasa']),
             textcoords="offset points", xytext=(0,10), ha='center', color='red')

plt.xlabel('Año')
plt.ylabel('Tasa de desempleo (%)')
plt.title('Tasa de desempleo anual por género')
plt.legend()
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

```

 Podemos plantear algunas hipótesis sobre los cambios en la TDA (Tasa de Desocupación Abierta) en estos años. 


Evidentemente, en el año 2020 debido a la pandemia la tasa de desempleo se elevó exponencialmente, 
llegando hasta 17.9% en el caso de las mujeres y 16.2% en el caso de los hombres.
