```python
#Cambiamos el nombre de las columnas por la primera fila
grad_pub2.columns = grad_pub2.iloc[1]
grad_pub2.drop(grad_pub2.index[0:3],inplace=True)
```

```python
grad_pub2.head(4)
```

```python
# Primero reseteamos los index
grad_pub2.reset_index(inplace=True, drop=True)
```

```python
grad_pub2.head(4)
```

```python
# Los valores de la columna NaN deberían ir en Universidades Públicas
# Para lograr esto, eliminaremos la columna Universidades Públicas y
# renombraremos NaN

grad_pub2.drop("Universidades públicas", axis=1, inplace=True)
# La columna sin nomnbre es NaN
grad_pub2.columns = grad_pub2.columns.fillna('Universidades públicas')

# Eliminamos fila vacía
grad_pub2.drop(1,axis=0,inplace=True)
```

```python
grad_pub2.head(5)
```

```python
# Eliminamos las filas donde todos los valores sean NaN
grad_pub2=grad_pub2.dropna(how="all", axis=0)
```

```python
grad_pub2.isna().sum()
```

```python
import numpy as np
np.sum(grad_pub2=="-")
```

```python
# Reemplazamos "-" con 0
grad_pub2 = grad_pub2.replace('-',0)
np.sum(grad_pub2=='-')
```

```python
# Obtenemos todos los graduados por año
grad_pub2=grad_pub2.reset_index(drop=True)
grad_pub2 = grad_pub2.rename_axis(None, axis=1)
# Convertir los años a int

grad_pub2.columns = ["uni"] + list(map(int, grad_pub2.columns[1:]))
grad_pub2.head(10)
```

```python
# Finalmente transponemos la df.

grad_pub2 = grad_pub2.set_index("uni").transpose()
```

```python
# Arreglamos los indices
grad_pub2=grad_pub2.reset_index()
grad_pub2= grad_pub2.rename(columns={'index':'año'})
grad_pub2 = grad_pub2.rename_axis(None, axis=1)
grad_pub2
```

```python
# Ahora generalizemos en una función para las universidades privadas
def cargar_uni(df):
    # Crear una copia del DataFrame original
    df_copy = df.copy()
    
    # Renombrar columnas y limpiar datos
    df_copy.columns = df_copy.iloc[1]
    df_copy.drop(df_copy.index[0:3], inplace=True)
    df_copy.drop("Universidades privadas", axis=1, inplace=True)
    
    # Renombrar columna NaN
    df_copy.columns = df_copy.columns.fillna('Universidades privadas')
    
    # Resetear índice
    df_copy.dropna(how="all", axis=0, inplace=True)
    df_copy.replace('-', 0, inplace=True)
    df_copy.reset_index(drop=True, inplace=True)
    df_copy = df_copy.rename_axis(None, axis=1)
    
    # Convertir los años a int
    df_copy.columns = ["uni"] + list(map(int, df_copy.columns[1:]))
    df_copy["uni"] = df_copy["uni"].str.strip()

    df_copy=df_copy.set_index("uni").transpose()
    df_copy=df_copy.reset_index()
    df_copy= df_copy.rename(columns={'index':'año'})
    df_copy = df_copy.rename_axis(None, axis=1)

    return df_copy
    
    
```

```python
grad_priv2 = cargar_uni(grad_priv)
```

```python
grad_priv2.head(10)
```

```python
#Podemos visualizar los n° de graduados a lo largo del tiempo (no hay data del 2008 al 2011)

plt.figure(figsize=(10, 6))
plt.bar(grad_priv2["año"], grad_priv2["Universidad del Pacífico"], color='skyblue')
plt.title('Número de Graduados de la Universidad del Pacífico (2004-2022)')
plt.xlabel('Año')
plt.ylabel('Número de Graduados')
plt.xticks(range(2004,2023))
for i, txt in enumerate(grad_priv2["Universidad del Pacífico"]):
    plt.annotate(txt, (grad_priv2["año"][i], grad_priv2["Universidad del Pacífico"][i]), textcoords="offset points", xytext=(0,10), ha='center')

plt.grid(True)
plt.tight_layout()
```

```python
# Ahora veamos el nivel de educación alcanzado por la población de 15 años a más

nivel = pd.read_excel('cd4_35.xlsx')
```

```python
nivel.head(15)
```

```python
# Formateamos las columnas como en el caso anterior

def cargar(df):
    # Crear una copia del DataFrame original
    df_copy = df.copy()
    df_copy.drop(df_copy.index[11:], inplace=True)

    # Renombrar columnas y limpiar datos
    df_copy.columns = df_copy.iloc[3]
    df_copy.drop(df_copy.index[0:4], inplace=True)

    # Convertir los años a int
    df_copy.columns = ["nivel"] + list(map(int, df_copy.columns[1:]))
    df_copy.dropna(how='all', axis=0, inplace=True)
    df_copy=df_copy.reset_index(drop=True)
    df_copy = df_copy.rename_axis(None, axis=1)
    
    

    return df_copy
    
    
```

```python
df_copy=cargar(nivel)
```

```python
plt.figure(figsize=(10,6))
for level in df_copy['nivel'].unique():
    df_level = df_copy[df_copy['nivel'] == level]
    plt.plot(df_level['year'], df_level['count'], marker='o', linestyle='-', label=level)

plt.title('Nivel de Educación (2008-2022)')
plt.xlabel('Año')
plt.ylabel('Número de Personas')
plt.grid(True)
plt.legend(title='Nivel de Educación')
plt.xticks(range(2008, 2023), rotation=45)
plt.tight_layout()

# Show the plot
plt.show()
```

```python
df_copy2=df_copy.set_index("nivel")
df_copy2=df_copy2.drop("Total")
df_copy2=df_copy2.T
df_copy2.rename(columns={"Primaria":"año"})
```

```python
df_copy2=df_copy.set_index("nivel").T
df_copy2=df_copy2.drop("Total",axis=1)
df_copy2=df_copy2.rename_axis(None,axis=1)
df_copy2=df_copy2.reset_index()
df_copy2=df_copy2.rename(columns={"index":"año"})
```

```python
df_copy2.columns = df_copy2.columns.str.strip()
```

```python
df_copy2.columns
```

```python
df_copy2.iloc[:, 1:] = df_copy2.iloc[:, 1:].apply(pd.to_numeric, errors='coerce')
```

```python
df_copy2.set_index('año', inplace=True)
df_copy2.plot(kind='bar', stacked=True, figsize=(12, 8))
plt.title('Distribución de Niveles Educativos (2008-2022)')
plt.xlabel('Año')
plt.ylabel('Porcentaje')
plt.legend(title='Nivel Educativo', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(12, 8))
for column in df_copy2.columns[1:]:
    plt.plot(df_copy2.index, df_copy2[column], marker='o', label=column)
plt.title('Tendencia de Niveles Educativos (2008-2022)')
plt.xlabel('Año')
plt.ylabel('Porcentaje')
plt.legend(loc='upper left')
plt.grid(True)
plt.show()
```

```python

# Número de filas y columnas para los subplots
num_years = df_copy2.shape[0]
cols = 3
rows = (num_years // cols) + (num_years % cols > 0)

# Crear subplots
fig, axes = plt.subplots(rows, cols, figsize=(15, 15))
fig.suptitle('Distribución de Niveles Educativos por Año', fontsize=16)

# Aplanar los ejes para facilitar la iteración
axes = axes.flatten()

# Iterar sobre cada año y crear un gráfico de pastel
for i, (year, row) in enumerate(df_copy2.iterrows()):
    ax = axes[i]
    row.plot(kind='pie', ax=ax, autopct='%1.1f%%', startangle=90, legend=False)
    ax.set_title(f'Año {year}')
    ax.set_ylabel('')

# Eliminar ejes vacíos si existen

plt.tight_layout()
plt.subplots_adjust(top=0.95)
plt.show()
```
